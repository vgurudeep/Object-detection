```python
import cv2 
import matplotlib.pyplot as plt
```

# Importing the pretrained model( mobilenet.v3) to our memory

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn.DetectionModel(frozen_model, config_file) ## loading our model
```

## Some image preprocessing before we pass it to our model

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)  ## 255 is the max so 255/2= 127.5
model.setInputMean((127.5,127.5,127.5)) ## mobilenet ==> [-1,1]
model.setInputSwapRB(True) ## converting bgr to rgb
```

```python
classLabels = [] ## creating an emypty list
file_name = 'Labels.txt' 
with open(file_name, 'rt') as fpt: ## copying file contents to the empty list
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
classLabels
```

```python
img = cv2.imread('man-tuxedo-car-style-wallpaper-preview.jpg') ## reading the image
```

```python
plt.imshow(img) ## displaying image
```

## Passing our input image to the model

```python
ClassIndex, confidence, bbox = model.detect(img, confThreshold=0.5)  
```

```python
ClassIndex
```

```python
confidence
```

```python
bbox
```

## For Image

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(),confidence.flatten(),bbox):
    cv2.rectangle(img,boxes,(0,0,127),1) ## drawing boxex around the object // image localization
    cv2.putText(img, classLabels[ClassInd-1], (boxes[0]+10,boxes[1]+40),font, fontScale=font_scale, color=(255,0,0), thickness=3) ## labelling the boxes
```

```python
plt.imshow(img)
```

## For Video capture

```python
cap = cv2.VideoCapture('The CCTV People Demo 2.mp4')


if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError('Cannot open video')
    
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret,frame = cap.read()
    
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.5)
    
    print(ClassIndex)
    if (len(ClassIndex)!=0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(),confidence.flatten(),bbox):
           if (ClassInd<=80):
            cv2.rectangle(frame,boxes,(0,0,127),1)
            cv2.putText(frame, classLabels[ClassInd-1], (boxes[0]+10,boxes[1]+40),font, fontScale=font_scale, color=(255,0,0), thickness=3)
    cv2.imshow('boject detection', frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()


```
